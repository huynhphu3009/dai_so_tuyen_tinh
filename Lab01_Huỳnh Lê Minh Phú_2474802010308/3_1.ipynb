```python
import numpy as np

#Khởi tạo vector: vector1
vec1= np.array([1, 3, 5])
print("vector1:", vec1)

#Nhân vector1 với 2
vec2= vec1 * 2
print("vec2:", vec2)

#Nhân vector1 với chính nó
vec3= vec1 * vec1
print("vec3:", vec3)

#Chia vector1 cho 2
vec4= vec1 / 2
print("vec4:", vec4)

#Tạo vector giả trị rỗng (0) dài 5 giá trị
vec5= np.zeros(5)
print("vec5 (zeros):", vec5)

#Tạo vector ngẫu nhiên từ 0 đến 1, kích thước 5
vec6= np.ones(5)
print("vec6 (ones):", vec6)

#Tạo vector ngẫu nhiên từ 0 đến 1, kích thước 5, giá trị ngẫu nhiên
vec7= np.random.rand(5)
print("vec7 (random):", vec7)

#Lấy phần tử thứ 3 của vec3
print("Phần tử thứ 3 của vec3:", vec3[2])

#Tách phần tử từ cuối cùng của vec4
print("Phần tử cuối của vec4:", vec4[-1])

#Sửa phần tử thứ 2 của vec1 thành 10
vec1[1]= 10
print("vec1 sau khi chỉnh sửa:", vec1)

#Tìm căn bậc hai từng phần tử của vector
sqrt_vec= np.sqrt(vec1)
print("Căn bậc hai của vec1:", sqrt_vec)

#Tìm cos của từng phần tử của vector
cos_vec= np.cos(vec1)
print("Cos của vec1:", cos_vec)

#Tìm sin của từng phần tử của vector
sin_vec= np.sin(vec1)
print("Sin của vec1:", sin_vec)

#Tích vô hướng của hai vector
dot_product= np.dot(vec1, vec3)
print("Tích vô hướng của vec1 và vec3:", dot_product)

#Véc-tơ hình dạng (shape)
print("Shape của vec1:", vec1.shape)

#Tinh chuyển vị vector (transpose)
#Với vector 1D, transpose không thay đổi
transposed_vec1= np.transpose(vec1)
print("Chuyển vị của vec1:", transposed_vec1)

#Tinh vecto ẩn danh (ấn định giá trị ngẫu nhiên)
vec8= np.random.random()
print("Vector ẩn danh (giá trị ngẫu nhiên):", vec8)

#Tạo vector toàn 0
vec9= np.zeros(5)
print("Vector toàn 0:", vec9)

#Tạo vector ngẫu nhiên từ 0 đến 1 cảnh báo: np.random.random() đã làm
vec10= np.random.rand(5)
print("Vector ngẫu nhiên có phần tử từ 0 đến 1:", vec10)

#Phép cộng của các vector
sum_vec= np.sum([vec1, vec2, vec3], axis=0)
print("Tổng các vector:", sum_vec)

#Thử nghiệm chuyển vị vector
v_transpose= np.transpose(vec2)
print("Chuyển vị của vec2:", v_transpose)

#Sử dụng phép gán, ví dụ: gán giá trị mới cho vec1
vec1= np.array([4, 6])
print("vec1 mới:", vec1)

#Ví dụ về phép cộng matrix-vector
matrix= np.array([[1, 2], [3, 4]])
vector= np.array([5, 6])
result= np.dot(matrix, vector)
print("Kết quả nhân matrix với vector:", result)
```
